```python
import pandas as pd 
import numpy as np
```

```python
dt = pd.read_csv('customer-churn.csv')
```

```python
# create fiber optics feature #
OH_internetService = pd.get_dummies(dt['InternetService'])
```

```python
dt = pd.concat([OH_internetService,dt],axis=1)
```

```python
dt.to_csv('processed_data.csv',index=False)
```

```python
# create support flag #
OH_Techsupport = pd.get_dummies(dt['TechSupport'],prefix='TechSupport')
```

```python
OH_Techsupport
```

```python
dt = pd.concat([OH_Techsupport,dt],axis=1)
```

```python
dt.to_csv('processed_data.csv',index=False)
```

```python
# all service flag #
def all_services(row):
    if row[0] == 'Yes' and row[1]=='Yes' and row[2] == 'Yes' and row[3]=='Yes' and row[4] == 'Yes' and row[5]=='Yes' and row[6] == 'Yes' and row[7]=='Yes':
        return 1
    else:
        return 0
```

```python
services = ['PhoneService', 'MultipleLines',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies']
```

```python
dt['All services flag'] = dt[services].apply(lambda x: all_services(x),axis=1)
```
